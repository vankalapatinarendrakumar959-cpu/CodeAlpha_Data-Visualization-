## Objective:
Transform raw sales data into meaningful visualizations
using Python libraries like Matplotlib and Seaborn.

## Tools Used:
- Python
- Pandas
- Matplotlib
- Seaborn
- Google Colab

## Dataset Description

The dataset contains monthly business performance data:

Columns:
- Month: Sales month
- Sales: Total sales amount
- Customers: Number of customers
- Profit: Monthly profit

The goal is to analyze trends and identify business insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style
sns.set_style("whitegrid")

print("Libraries imported successfully")
```

```python
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun"],
    "Sales": [25000,30000,28000,45000,50000,60000],
    "Customers": [120,150,140,200,240,300],
    "Profit": [5000,7000,6500,12000,15000,20000]
}

df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(8,5))

plt.plot(df["Month"], df["Sales"], marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(df["Month"], df["Profit"])

plt.title("Monthly Profit Analysis")
plt.xlabel("Month")
plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=df["Month"],
    y=df["Customers"]
)

plt.title("Monthly Customer Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x=df["Sales"],
    y=df["Profit"],
    s=100
)

plt.title("Sales vs Profit Relationship")

plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,8))

# Sales
axes[0,0].plot(df["Month"], df["Sales"], marker="o")
axes[0,0].set_title("Sales Trend")

# Profit
axes[0,1].bar(df["Month"], df["Profit"])
axes[0,1].set_title("Profit Analysis")

# Customers
axes[1,0].bar(df["Month"], df["Customers"])
axes[1,0].set_title("Customer Growth")

# Sales vs Profit
axes[1,1].scatter(df["Sales"], df["Profit"])
axes[1,1].set_title("Sales vs Profit")

plt.tight_layout()

plt.savefig("Business_Dashboard.png")

plt.show()
```

## Data Insights

1. Sales increased continuously from January to June.
2. Profit growth followed the sales trend.
3. Customer count increased every month.
4. Sales and profit have a positive relationship.
5. The best performing month was June.
